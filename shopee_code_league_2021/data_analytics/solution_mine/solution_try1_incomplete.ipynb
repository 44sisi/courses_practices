```python
import json
import pandas as pd
import numpy as np
import networkx as nx
```

```python
df = pd.read_json("contacts.json")
df.head(20)
```

```python
df.loc[[6, 38, 32871, 142067, 236367]]
```

```python
df.replace("", np.nan, inplace=True)
df
```

```python
df_email = df[['Id','Email']]
df_email.dropna(subset=['Email'], inplace=True)
df_email = df_email.set_index('Email').astype(str).groupby('Email').agg(',' .join).reset_index()
df_email["Id"] = df_email["Id"].apply(lambda x: sorted(x.split(",")))
df_email
```

```python
email_id = df_email["Id"].tolist()
# email_id = list(filter(lambda x: len(x) > 1, email_id))
print(len(email_id))
email_id[:10]
```

```python
df_phone = df[['Id','Phone']]
df_phone.dropna(subset=['Phone'], inplace=True)
df_phone = df_phone.set_index('Phone').astype(str).groupby('Phone').agg(',' .join).reset_index()
df_phone["Id"] = df_phone["Id"].apply(lambda x: sorted(x.split(",")))
df_phone
```

```python
phone_id = df_phone["Id"].tolist()
# phone_id = list(filter(lambda x: len(x) > 1, phone_id))
print(len(phone_id))
phone_id[:10]
```

```python
df_orderid = df[['Id','OrderId']]
df_orderid.dropna(subset=['OrderId'], inplace=True)
df_orderid = df_orderid.set_index('OrderId').astype(str).groupby('OrderId').agg(',' .join).reset_index()
df_orderid["Id"] = df_orderid["Id"].apply(lambda x: sorted(x.split(",")))
df_orderid
```

```python
orderid_id = df_orderid["Id"].tolist()
# orderid_id = list(filter(lambda x: len(x) > 1, orderid_id))
print(len(orderid_id))
orderid_id[:10]
```

```python
nodes = email_id + phone_id + orderid_id
nodes[:10]
```

```python
import networkx 
from networkx.algorithms.components.connected import connected_components

def to_graph(l):
    G = networkx.Graph()
    for part in l:
        # each sublist is a bunch of nodes
        G.add_nodes_from(part)
        # it also imlies a number of edges:
        G.add_edges_from(list(zip(part[:-1], part[1:])))
    return G  

G = to_graph(nodes)
connected_ids = list(connected_components(G))
connected_ids[:10]
```
